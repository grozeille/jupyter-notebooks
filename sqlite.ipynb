```python
import numpy as np
import pandas as pd
listings_df = pd.read_csv("listings.csv") 
listings_df.head()
```

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite:///db.sqlite', echo=False)
```

```python
engine.execute('DROP TABLE IF EXISTS listings')
```

```python
# import csv to db
listings_df.to_sql('listings', con=engine, index=False, chunksize=1000, if_exists='replace')
```

```python
sql_DF = pd.read_sql("SELECT * FROM listings where demand > 0 limit 10", con=engine)
sql_DF
```

```python
engine.execute('DROP TABLE IF EXISTS commodities')
```

```python
commoditites_df = pd.read_json (r'commodities.json')
# extract the name of the category object
commoditites_df['category_name'] = commoditites_df.apply(lambda row: row.category['name'], axis = 1) 
# remove the category column, because we can't save complexe objects in a database
commoditites_df = commoditites_df.drop(columns=['category'])
# save all into the database
commoditites_df.to_sql('commodities', con=engine, index=False, chunksize=1000, if_exists='replace')
```
